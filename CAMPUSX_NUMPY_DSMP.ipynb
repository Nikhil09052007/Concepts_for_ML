**Stacking** : Adding two or more arrays together


```python
#horinzontal stacking: It adds two arrays sidewise.
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
```

```python
np.hstack((a4,a5)) # can add as much as can
```

```python
# Vertical stacking: Adds from top to bottom of two arryas.

np.vstack((a4,a5))
```
