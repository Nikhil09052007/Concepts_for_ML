Numpy array vs Python List

```python
import numpy as np
```

```python
# Checking the effieciency of list and numpy arrayt
# for list

a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time
start = time.time()

for i in range(len(a)):
  c.append(a[i]+b[i])

end = time.time()

end-start
```

```python
# using numpy
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)

start = time.time()

c = a+b
end = time.time()

end-start
```
